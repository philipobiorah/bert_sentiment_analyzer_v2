### Pre Processing

```python
#Rmove html tags
def sanitize_text(input_text):
    # Remove HTML tags using regular expression
    clean_text = re.sub(r'<.*?>', '', input_text)
    
    #Remove special characters and exra spaces
    clean_text = re.sub(r'[^\w\s]', ' ', clean_text)
    
    #remove extra spaces and newline characters
    clean_text = ' '.join(clean_text.split())
    
    return clean_text
    
```

### Lexicon-based Approach:
>VADER relies on a predefined lexicon (a dictionary) of words and phrases, each of which is assigned a polarity score indicating how positive or negative it is. The lexicon includes words and phrases with sentiment, including slang, emoticons, and modifiers (e.g., "very" or "not").


>Valence Scores: Each word in the lexicon is assigned a sentiment score ranging from -1 (most negative) to 1 (most positive). For example, "happy" might have a positive score close to 1, while "sad" could have a negative score close to -1.

> Sentiment Intensity: VADER not only classifies text as positive, negative, or neutral but also provides a sentiment intensity score. This score indicates how strong the sentiment is in the text. A higher intensity score suggests a stronger sentiment.

> Handling Negation and Emphasis: VADER is designed to handle negations (e.g., "not good") and emphasizes (e.g., "very good") by considering the impact of modifiers and context on sentiment.

### Step 1. VADER Seniment Scoring

#Word2Vector

```python
from gensim.models import Word2Vec
from gensim.models.keyedvectors import KeyedVectors
import numpy as np

# Assuming `norm_train_reviews` and `norm_test_reviews` are lists of tokenized texts (lists of words)
# For example: norm_train_reviews = [["word1", "word2", ...], ["word3", "word4", ...], ...]

# Train a Word2Vec model on your training data
word2vec_model = Word2Vec(sentences=norm_train_reviews, vector_size=100, window=5, min_count=1, workers=4)

# Alternatively, you can load a pre-trained Word2Vec model if you have one
# word2vec_model = KeyedVectors.load_word2vec_format('path_to_pretrained_word2vec.bin', binary=True)

# Define a function to convert reviews to vectors
def document_vector(word2vec_model, doc):
    # remove out-of-vocabulary words
    doc = [word for word in doc if word in word2vec_model.wv]
    if len(doc) == 0:
        return np.zeros(word2vec_model.vector_size)
    return np.mean(word2vec_model.wv[doc], axis=0)

# Vectorize the training and testing reviews
train_vectors = np.array([document_vector(word2vec_model, review) for review in norm_train_reviews])
test_vectors = np.array([document_vector(word2vec_model, review) for review in norm_test_reviews])

print("Word2Vec_train:", train_vectors.shape)
print("Word2Vec_test:", test_vectors.shape)
```

### Roberta Pretrained Model

```python
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
```

```python
#using Pre-trained model
```

```python
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment"
r_tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
sia.polarity_scores(example)
```

```python
#Run for Roborto
encoded_text = r_tokenizer(example, return_tensors='pt')
output = model(**encoded_text)
```

```python
scores = output[0][0].detach().numpy()
```

```python
scores
```

```python
def polarity_scores_roberta(example):
    encoded_text = r_tokenizer(example,truncation=True,max_length=512, padding='max_length', return_tensors='pt')
    output = model(**encoded_text)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    scores_dict = {
        'roberta_neg' : scores[0],
        'roberta_neu' : scores[1],
        'roberta_pos' : scores[2]
    }
    return scores_dict
```

```python
# A dictionary to hold the Roberta results
roberta_result = {}
```

```python
for i, row in tqdm(train_df.iterrows(), total=len(train_df)):
    try:
        
        text = row['review']
        myid = row['ID']
        roberta_result[myid] = polarity_scores_roberta(text)
        
    except RuntimeError:
        print(f'Broke for id {myid}')
        
```

```python
# Define a function that takes a row of the DataFrame and returns the sentiment label
def get_sentiment_label(row):
    # Compare only the negative and positive scores
    if row['roberta_neg'] > row['roberta_pos']:
        return 'negative'
    else:
        return 'positive'

```

```python
# Now apply this function to each row of the DataFrame to create a new column 'predicted_sentiment'
roBERTa['predicted_sentiment'] = roBERTa.apply(get_sentiment_label, axis=1)
```
