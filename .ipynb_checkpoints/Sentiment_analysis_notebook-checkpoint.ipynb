### Pre Processing

```python
#Rmove html tags
def sanitize_text(input_text):
    # Remove HTML tags using regular expression
    clean_text = re.sub(r'<.*?>', '', input_text)
    
    #Remove special characters and exra spaces
    clean_text = re.sub(r'[^\w\s]', ' ', clean_text)
    
    #remove extra spaces and newline characters
    clean_text = ' '.join(clean_text.split())
    
    return clean_text
    
```

```python
train_df['review'] = train_df['review'].apply(sanitize_text)
test_df['review'] = test_df['review'].apply(sanitize_text)
```

### Lexicon-based Approach:
>VADER relies on a predefined lexicon (a dictionary) of words and phrases, each of which is assigned a polarity score indicating how positive or negative it is. The lexicon includes words and phrases with sentiment, including slang, emoticons, and modifiers (e.g., "very" or "not").


>Valence Scores: Each word in the lexicon is assigned a sentiment score ranging from -1 (most negative) to 1 (most positive). For example, "happy" might have a positive score close to 1, while "sad" could have a negative score close to -1.

> Sentiment Intensity: VADER not only classifies text as positive, negative, or neutral but also provides a sentiment intensity score. This score indicates how strong the sentiment is in the text. A higher intensity score suggests a stronger sentiment.

> Handling Negation and Emphasis: VADER is designed to handle negations (e.g., "not good") and emphasizes (e.g., "very good") by considering the impact of modifiers and context on sentiment.

### Machine Learning Apporoach : 

#### Stemming

```python
def text_stemmer(text):
    ps=nltk.porter.PorterStemmer()
    text = ' '.join([ps.stem(word) for word in text.split()])
    return text
    
```

```python
train_df['review'] = train_df['review'].apply(text_stemmer)
test_df['review'] = test_df['review'].apply(text_stemmer)
```

```python
### Stop Words Removal
def remove_stopwords(text):
    # Tokenize the text into words
    words = nltk.word_tokenize(text)
    
    #Remove stop words from the text
    filtered_words = []
    for word in words:
        if word.lower() not in stopword_list:
            filtered_words.append(word)
            
    filtered_text = ' '.join(filtered_words) 
    return filtered_text
```

```python
#Apply function on reveiow column
train_df["review"] = train_df["review"].apply(remove_stopwords)
test_df['review'] = test_df['review'].apply(remove_stopwords)
```

```python
norm_train_reviews = train_df.review[:-1]
norm_train_reviews
```

```python
norm_test_reviews = test_df.review
norm_test_reviews
```

```python
#### Bags of words model
# We convert text document to numerical vectors or bag of words
```

```python
#We use count vectorizer for bag of words
cv = CountVectorizer(min_df=0.0, max_df=1.0, binary=False, ngram_range=(1,3))
# Fit and transform the training reviews to create a count matrix
cv_train_reviews = cv.fit_transform(norm_train_reviews)
```

```python
# Transform the test reviews using the vocabulary learned from training
cv_test_reviews = cv.transform(norm_test_reviews)
```

```python
print("BOW_cv_train:", cv_train_reviews.shape)
print("BOW_cv_test:", cv_test_reviews.shape)
```
