#### Data Cleaning

```python
#Rmove html tags
def sanitize_text(input_text):
    # Remove HTML tags using regular expression
    clean_text = re.sub(r'<.*?>', '', input_text)
    
    #Remove special characters and exra spaces
    clean_text = re.sub(r'[^\w\s]', ' ', clean_text)
    
    #remove extra spaces and newline characters
    clean_text = ' '.join(clean_text.split())

     # Convert to lower case
    clean_text = clean_text.lower()

    #Remove stopwords
    words = clean_text.split()
    stops = set(stopwords.words("english"))
    clean_text  = [w for w in words if not w in stops]
    
    # Join the words back into one string
    return " ".join(clean_text)
```

```python
df['review'] = df['review'].apply(sanitize_text)
```

### Feature Extraction

#### Using Word2Vec

```python
# Tokenize the sentences
tokenized_sentences = [sanitize_text(review).split() for review in df['review']]
```

```python
# Train a Word2Vec model
word2vec_model = Word2Vec(tokenized_sentences, vector_size=100, window=5, min_count=2, workers=4)
```

```python
# Function to create a feature vector for each review
def get_feature_vector(words, model):
    feature_vector = np.zeros((model.vector_size,), dtype="float32")
    nwords = 0
    word_set = set(model.wv.index_to_key)
    
    for word in words:
        if word in word_set:
            feature_vector = np.add(feature_vector, model.wv[word])
            nwords += 1
            
    if nwords:
        feature_vector = np.divide(feature_vector, nwords)
    return feature_vector
```

```python
# Create feature vectors for each review
word2vec_features = np.array([get_feature_vector(review, word2vec_model) for review in tokenized_sentences])
```

### Model Training and Evaluation

```python
# Function to train and evaluate model
def train_evaluate(features, labels):
    X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
    
    # Train the Logistic Regression model
    model = LogisticRegression()
    model.fit(X_train, y_train)
    
    # Predictions
    predictions = model.predict(X_test)
    
    # Confusion matrix
    cm = confusion_matrix(y_test, predictions)
    # Classification report

    # Classification report
    cr = classification_report(y_test, predictions, output_dict=True)

    return cm, cr
    
```

```python
# Convert sentiments to binary
df['sentiment'] = df['sentiment'].map({'positive': 1, 'negative': 0})
```

### VADER (Valence Aware Dictionary and sEntiment Reasoner) 

```python
#import required libraries
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
# Download the VADER lexicon
nltk.download('vader_lexicon')

# Initialize the VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()
```

```python
# Function to apply VADER and get compound score
def vader_sentiment(review):
    return sia.polarity_scores(review)["compound"]

# Apply the function to the review column
df['vader_score'] = df['review'].apply(vader_sentiment)
```

```python
df
```

VADER's Compound Score: VADER generates a "compound" score for a given text, which is a normalized, weighted composite score. This score ranges from -1 (most extreme negative) to +1 (most extreme positive).

Threshold Interpretation:

A positive sentiment is usually indicated by a compound score greater than 0.05.
A negative sentiment is indicated by a compound score less than -0.05.
Scores between -0.05 and 0.05 are typically considered neutral.

```python
#convert sentiment to binaray 
# Function to convert VADER score to binary class
def to_binary_class(score, threshold=0.05):
    return 1 if score > threshold else 0

# Apply the function to the Vader scores
df['predicted_sentiment'] = df['vader_score'].apply(lambda x: to_binary_class(x))
```

## Modeling

### Lexicon-based Approach:
>VADER relies on a predefined lexicon (a dictionary) of words and phrases, each of which is assigned a polarity score indicating how positive or negative it is. The lexicon includes words and phrases with sentiment, including slang, emoticons, and modifiers (e.g., "very" or "not").


>Valence Scores: Each word in the lexicon is assigned a sentiment score ranging from -1 (most negative) to 1 (most positive). For example, "happy" might have a positive score close to 1, while "sad" could have a negative score close to -1.

> Sentiment Intensity: VADER not only classifies text as positive, negative, or neutral but also provides a sentiment intensity score. This score indicates how strong the sentiment is in the text. A higher intensity score suggests a stronger sentiment.

> Handling Negation and Emphasis: VADER is designed to handle negations (e.g., "not good") and emphasizes (e.g., "very good") by considering the impact of modifiers and context on sentiment.

### Step 1. VADER Seniment Scoring
> This uses a "bag of words" approach: Stop words are removed

> each word is scored and combined to a total score.

```python
from gensim.models import Word2Vec
from gensim.models.keyedvectors import KeyedVectors
import numpy as np

# Assuming `norm_train_reviews` and `norm_test_reviews` are lists of tokenized texts (lists of words)
# For example: norm_train_reviews = [["word1", "word2", ...], ["word3", "word4", ...], ...]

# Train a Word2Vec model on your training data
word2vec_model = Word2Vec(sentences=norm_train_reviews, vector_size=100, window=5, min_count=1, workers=4)

# Alternatively, you can load a pre-trained Word2Vec model if you have one
# word2vec_model = KeyedVectors.load_word2vec_format('path_to_pretrained_word2vec.bin', binary=True)

# Define a function to convert reviews to vectors
def document_vector(word2vec_model, doc):
    # remove out-of-vocabulary words
    doc = [word for word in doc if word in word2vec_model.wv]
    if len(doc) == 0:
        return np.zeros(word2vec_model.vector_size)
    return np.mean(word2vec_model.wv[doc], axis=0)

# Vectorize the training and testing reviews
train_vectors = np.array([document_vector(word2vec_model, review) for review in norm_train_reviews])
test_vectors = np.array([document_vector(word2vec_model, review) for review in norm_test_reviews])

print("Word2Vec_train:", train_vectors.shape)
print("Word2Vec_test:", test_vectors.shape)
```
